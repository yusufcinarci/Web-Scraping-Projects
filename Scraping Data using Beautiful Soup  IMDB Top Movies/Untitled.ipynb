# import Modules

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

# Request page source from URL

```python
url="https://www.imdb.com/chart/top/"
```

```python
page=requests.get(url)
page
```

```python
#display the page source code
page.content
```

```python
soup = BeautifulSoup(page.content,"html.parser")
print(soup.prettify())
```

```python
scrapped_movies = soup.find_all('td',class_='titleColumn')
scrapped_movies
```

```python
#parse movie names
movies = []
for movie in scrapped_movies:
    movie = movie.get_text().replace('\n',"")
    movie=  movie.strip(" ")
    movies.append(movie)
movies
```

```python
scrapped_ratings = soup.find_all('td',class_='ratingColumn imdbRating')
scrapped_ratings
```

```python
#parse movie names
ratings = []
for rating in scrapped_ratings:
    rating = rating.get_text().replace('\n',"")
    rating =  rating.strip(" ")
    ratings.append(rating)
ratings
```

# Store the Scraped Data

```python
data=pd.DataFrame()
data['Movie Names']=movies
data['Ratings']=ratings
data.head()
```

```python
data.to_csv('IMDB Top Movies.csv',index = False)
```
